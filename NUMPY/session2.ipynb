# Numpy arrays vs python list

### speed:

```python
#list:

import time

a = [i for i in range(10000000)]
b = [i for i in range(10000000, 20000000)]

c = []

start = time.time()

for i in range(len(a)):
    c.append(a[i] + b[i])

print(time.time() - start)
```

```python
#numpy arrays:

import numpy as np

a = np.arange(10000000)
b = np.arange(10000000,20000000)

start = time.time()
c=a+b

print(time.time() - start)
```

#### memory space:

```python
#list:

a = [i for i in range(10000000)]

import sys

sys.getsizeof(a)
```

```python
#numpy arrays:

a = np.arange(10000000)

sys.getsizeof(a)
```

```python
#also in numpy arrays we can do dtype!!

a = np.arange(10000000, dtype=np.int32)
sys.getsizeof(a)
```

### conveneience

```python
#list:

l1 = [1,2,3,4,5]
l2 = [6,7,8,9,10]

c = []

for i in range(len(l1)):
    c.append(l1[i] + l2[i])

print(c)
```

```python
#numpy arrays:

a1 = np(1,2,3,4,5)
a2 = np(6,7,8,9,10)

c = a1+a2

print(c)
```
