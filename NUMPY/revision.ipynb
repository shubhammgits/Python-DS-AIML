## creating a numpy array

```python
a = np.array([1,2,3,4])
print(a)
print(type(a))
```

```python
#2d ndarray : matrics

b = np.array([[1,2,3],[4,5,6]])
print(b)
```

```python
#3d nd array: tensor

c= np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(c)
```

```python
#aanother datatype:

a = np.array([1,2,3,4], dtype=float)
print(a)
print(type(a))
print(type(a[0]))
```

```python
#np.arange()

a = np.arange(1,6)
print(a)
```
