```python
import numpy as np
import random

arr1 = np.random.randint(1,100,24).reshape(6,4)
print(arr1)
```

```python
print(arr1[arr1>50])
```

```python
print(arr1[arr1%2 == 0])
```

```python
print(arr1[(arr1>50) & (arr1%2 == 0)])
```
