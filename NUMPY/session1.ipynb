# NUMPY

#### Creating numpy arrays:

```python
import numpy as np

a = np.array([1,2,3])
print(a)
```

```python
#2D and 3D numpy arrays:

b = np.array([[1,2,3], [4,5,6]])
print(b)
```

```python
c = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(c)
```

#### dtype

```python
np.array([1,2,3], dtype=float)
```

```python
np.array([1,2,3], dtype=bool)
```

```python
np.array([1,2,3], dtype=complex)
```

#### np.arange()

#### np.identity()

### Array Attributes

```python
a1 = np.arange(10)
a2 = np.arange(12, dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

#### ndim

```python
print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
```

#### shape

```python
print(a1.shape)
print(a2.shape)
print(a3.shape)
```

#### size

```python
print(a1.size)
print(a2.size)
print(a3.size)
```

#### itemsize

```python
print(a1.itemsize)
print(a2.itemsize)
print(a3.itemsize)
```
