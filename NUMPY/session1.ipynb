## Indexing and Slicing

```python
a1 = np.arange(10)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

```python
#Indexing:

#1d array:

print(a1[-1])
```

```python
#2d array indexing:

print(a2[0][3])
print(a2[2][1])

print(a2[1,0])
```

```python
#3d array:

print(a3[0,0,0])
print(a3[0,1,1])
```

```python
#Slicing

#1d array:

print(a1[3:7])
print(a1[::-1])
```

```python
#2d array:

print(a2[0,:])
print(a2[:,2])

print(a2[1:,1:3])

print(a2[::2,::3])
```

```python
#3d array:

a3 = np.arange(27).reshape(3,3,3)
print(a3)

```

```python
a3[::2,0,::2]
```
