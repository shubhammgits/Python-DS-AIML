# np.sort

```python
import numpy as np
a = np.random.randint(1,100,15)
print(a)
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
print(b)
```

```python
print(np.sort(a))
```

```python
print(np.sort(b))
```

```python
print(np.sort(b,axis=0))
```

```python
print(np.sort(a)[::-1])
```

# np.append

```python
a = np.arange(5)
np.append(a,200)
```

```python
b = np.arange(1,13).reshape(4,3)

np.append(b, np.ones((b.shape[0], 1)), axis=1)
```

# np.concatenate

```python
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

np.concatenate((c,d), axis=0)
```

```python
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

np.concatenate((c,d), axis=1)
```

# np.unique

```python
e = np.array([1,1,1,2,2,2,3,3,3,4,4,4,5,5,5,6,6])
f = np.array([[1,2,3,1], [1,2,3,2]])
```

```python
print(np.unique(e))
```

```python
print(np.unique(f))
```

## np.where

```python
a = np.random.randint(1,100,10)
print(a)
```

```python
#find all indices with value greater than 50
print(np.where(a>50))
```

```python
#replace all the vlues>50 with 0

np.where(a>50, 0, a)
```
