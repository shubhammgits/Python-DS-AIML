## np.where

```python
a = np.random.randint(1,100,10)
print(a)
```

```python
#find all indices with value greater than 50
print(np.where(a>50))
```

```python
#replace all the vlues>50 with 0

np.where(a>50, 0, a)
```

```python
#replace all even numbers with 0:

np.where(a%2==0, 0, a)
```
