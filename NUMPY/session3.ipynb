# np.sort

```python
import numpy as np
a = np.random.randint(1,100,15)
print(a)
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
print(b)
```

```python
print(np.sort(a))
```

```python
print(np.sort(b))
```

```python
print(np.sort(b,axis=0))
```

```python
print(np.sort(a)[::-1])
```
