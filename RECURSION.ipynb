```python
#eg: if we dont have any mulitplication operator(*)

#using loops:

def mult(a,b):
    result = 0

    for i in range(b):
        result += a

    print(result)

mult(3,5)
```

```python
#using recusrion:

def mult(a,b):
    if b==1:
        return a
    
    else:
        return a + mult(a,b-1)
    
print(mult(3,7))
```

```python
#factorial:

def fact(num):
    if num == 1:
        return num
    
    else:
        return num*fact(num-1)
    
fact(4)
```

```python
#Palindrome:

str1 = 'abba'
flag = True

for i in range(0, len(str1)//2  ):

    if str1[i] != str1[len(str1) - i - 1 ]:
        flag = False
        print('Not a palindrome!')
        break

if flag == True:
    print('Palindrome')
```

```python
def palin(text):
    if len(text) <= 1:
        print('palindrome')

    else:
        if text[0] == text[-1]:
            palin(text[1,-1])

        else:
            print('Not a palindrome!')

print(palin('madam'))
```
