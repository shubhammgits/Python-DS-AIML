# Decorators and Namespaces

```python
#local and global:

a = 3

def temp():
    b=8
    print(b)

temp()
print(a)
```

```python
a = 3

def temp():
    a=8
    print(a)

temp()
print(a)
```

```python
a = 3

def temp():
    print(a)

temp()
print(a)
```

```python
a = 3

def temp():
    global a
    a+=1
    print(a)

temp()
print(a)
```

```python
def temp():
    global a
    a=1
    print(a)

temp()
print(a)
```

```python
#built-in scope: built-in functions:
print('hello')

import builtins
print(dir(builtins))
```

```python
L = [1,2,3]
print(max(L))

def max():
    print('hello')

max(L)
```

```python
#enclosed:

def outer():
    def inner():
        print('Inner function')
    
    inner()
    print('Outer function')

outer()
print('Main program!')
```

```python
def outer():
    a=3
    def inner():
        print(a)
    inner()

a=1
outer()
```

```python
def outer():
    a=3
    def inner():
        nonlocal a
        a+=1
        print(a)
    inner()

a=1
outer()
```
