# File Handling + Serialization and Deserialization

#### Writing to a file

```python
# Case 1 : If the file is not present(not created)

f = open('sample.txt', 'w')
f.write('Hello World')
f.close()
```

```python
# writing multiple lines:

f = open('sample1.txt', 'w')
f.write('Hello World')
f.write('\nHow are you?')
f.close()
```

```python
# Case 2 : If file already exist:

f = open('sample.txt', 'w')
f.write('Salman Khan')
f.close()
```

#### Append mode

```python
f = open('sample1.txt', 'a')
f.write('\nI am fine!')
f.close()
```

#### Write lines()

```python
L = ['Hello', '\nHi', '\nHow are you?', '\nI am fine.']

f = open('sample.txt', 'w')

f.writelines(L)
f.close()
```

#### Reading from a file

```python
#using read()

f = open('sample.txt', 'r')
s = f.read()
print(s)
f.close()
```

```python
#reading only some particular charcetrs

f = open('sample.txt', 'r')
s = f.read(10)
print(s)
f.close()
```

```python
#readline()

f = open('sample.txt', 'r')
print(f.readline(), end='')
print(f.readline(), end='')

f.close()
```

```python
#to not to use readline() again and again:

f = open('sample.txt', 'r')

while True:

    data = f.readline()

    if data == '':
        break
    else:
        print(data, end='')
```

#### with keyword

```python
with open('sample.txt', 'w') as f:
    f.write('selmon bhai')
```

```python
f.write('hello')
```

```python
with open('sample.txt', 'r') as f:
    print(f.read())
```

```python
with open('sample.txt', 'r') as f:
    print(f.read(10))
    print(f.read(10))
```

```python
#benfit: to load a big file in memory

big_L = ['hello world' for i in range(1000)]

with open('big.txt', 'w') as f:
    f.writelines(big_L)
```

```python
with open('big.txt', 'r') as f:

    chunk_size = 10

    while len(f.read(chunk_size)) > 0:

        print(f.read(chunk_size), end='***')
        f.read(chunk_size)
```

#### seek and tell

```python
with open('sample1.txt', 'r') as f:
    f.seek(10)
    print(f.read(10))
    print(f.tell())
```

```python
#seek during write:

with open('sample.txt', 'w') as f:
    f.write('Hello World')
    f.seek(0)
    f.write('Xa')
```

### Binary files

```python
#working with binary file: creating a copy of an existing binary file

with open('me.jpg', 'rb') as f:
    with open('me_copy.jpg', 'wb') as wf:
        wf.write(f.read())
```
