# File Handling + Serialization and Deserialization

#### Writing to a file

```python
# Case 1 : If the file is not present(not created)

f = open('sample.txt', 'w')
f.write('Hello World')
f.close()
```

```python
# writing multiple lines:

f = open('sample1.txt', 'w')
f.write('Hello World')
f.write('\nHow are you?')
f.close()
```

```python
# Case 2 : If file already exist:

f = open('sample.txt', 'w')
f.write('Salman Khan')
f.close()
```

#### Append mode

```python
f = open('sample1.txt', 'a')
f.write('\nI am fine!')
f.close()
```

#### Write lines()

```python
L = ['Hello', '\nHi', '\nHow are you?', '\nI am fine.']

f = open('sample.txt', 'w')

f.writelines(L)
f.close()
```

#### Reading from a file

```python
#using read()

f = open('sample.txt', 'r')
s = f.read()
print(s)
f.close()
```

```python
#reading only some particular charcetrs

f = open('sample.txt', 'r')
s = f.read(10)
print(s)
f.close()
```
