# File Handling + Serialization and Deserialization

#### Writing to a file

```python
# Case 1 : If the file is not present(not created)

f = open('sample.txt', 'w')
f.write('Hello World')
f.close()
```

```python
# writing multiple lines:

f = open('sample1.txt', 'w')
f.write('Hello World')
f.write('\nHow are you?')
f.close()
```

```python
# Case 2 : If file already exist:

f = open('sample.txt', 'w')
f.write('Salman Khan')
f.close()
```

#### Append mode

```python
f = open('sample1.txt', 'a')
f.write('\nI am fine!')
f.close()
```

#### Write lines()

```python
L = ['Hello', '\nHi', '\nHow are you?', '\nI am fine.']

f = open('sample.txt', 'w')

f.writelines(L)
f.close()
```

#### Reading from a file

```python
#using read()

f = open('sample.txt', 'r')
s = f.read()
print(s)
f.close()
```

```python
#reading only some particular charcetrs

f = open('sample.txt', 'r')
s = f.read(10)
print(s)
f.close()
```

```python
#readline()

f = open('sample.txt', 'r')
print(f.readline(), end='')
print(f.readline(), end='')

f.close()
```

```python
#to not to use readline() again and again:

f = open('sample.txt', 'r')

while True:

    data = f.readline()

    if data == '':
        break
    else:
        print(data, end='')
```

#### seek and tell

```python
with open('sample1.txt', 'r') as f:
    f.seek(10)
    print(f.read(10))
    print(f.tell())
```

```python
#seek during write:

with open('sample.txt', 'w') as f:
    f.write('Hello World')
    f.seek(0)
    f.write('Xa')
```

#### Problems with working with text file:

```python
#1. cant work with binary files like images using 'r' or 'w'ArithmeticError
```

```python
#working with binary file: creating a copy of an existing binary file

with open('me.jpg', 'rb') as f:
    with open('me_copy.jpg', 'wb') as wf:
        wf.write(f.read())
```

```python
#2. not good for other data types like int/float/list/tuples

with open('int.txt', 'w') as f:
    f.write(str(5))

with open('int.txt', 'r') as f:
    print(int(f.read()) + 5)
```

```python
#using complex data:

d = {
    'name' : 'Shubham',
    'age' : 19,
    'gender' : 'male'
}

with open('complex.txt', 'w') as f:
    f.write(str(d))

#we cannot perform operations on this dictionary as strings cant be converted into dictionary
```

### Serialization & Deserialization

```python
#serialization suing JSON module
#LIST:

import json

L = [1,2,3,4]

with open('demo.json', 'w') as f:
    json.dump(L,f)
```

```python
#dictionary:

d = {'name' : 'shubham', 'age' : 19, 'gender' : 'male'}

with open('demo.json', 'w') as f:
    json.dump(d,f, indent=4)
```

```python
#deserialization:

with open('demo.json', 'r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
#Tuple:

import json

t = (1,2,3,4,5)

with open('demo.json', 'w') as f:
    json.dump(t,f)

with open('demo.json', 'r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
#nested:

d = {
    'name' : 'shubham',
    'marks' : [34,56,32,89,100]
}

with open('demo.json', 'w') as f:
    json.dump(d,f)
```

### Custom objects serialization and deserialization

```python
class Person:
    def __init__(self, fname, lname, age, gender):
        self.fname = fname
        self.lname = lname
        self.age = age
        self.gender = gender

p = Person('Shubham', 'Jha', 20, 'Male')
```

```python
#serialization in form of dictionary:

def show_object(p):
    if isinstance(p, Person):
        return {
            'name' : p.fname + ' ' + p.lname, 'age' : p.age, 'gender' : p.gender
        }
    
with open('demo.json', 'w') as f:
    json.dump(p,f, default=show_object, indent=4)
```

```python
#deserialization:

with open('demo.json', 'r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```
