# LISTS

### Creation

```python
#empty:

L = []
print(L)
print(type(L))
```

```python
#1D
print([1,2,3,4])

#2D
print([1,2,3,[2,3,'hello']])

#3d
print([1,2,[2,3,[4,5]]])
```

```python
#Hetrogenous

print([1,2.3,'Hello', True])
```

```python
#type conversion
print(list('hello'))
```

### Accessing items from a list

```python
L = [1,2,[3,4]]

print(L[0])

print(L[-1])

print(L[-1][0])
```

```python
L = [[[1,2], [3,4]], [[5,6], [7,8]]]

print(L[1][0][0])
```

```python
l = [1,2,3,4,5,6]

print(l[0:3])
print(l[-3:])
print(l[::-1])
print(l[0::2])
```
