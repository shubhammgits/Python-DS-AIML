# LISTS

### Creation

```python
#empty:

L = []
print(L)
print(type(L))
```

```python
#1D
print([1,2,3,4])

#2D
print([1,2,3,[2,3,'hello']])

#3d
print([1,2,[2,3,[4,5]]])
```

```python
#type conversion
print(list('hello'))
```

### Accessing items from a list

```python
L = [1,2,[3,4]]

print(L[0])

print(L[-1])

print(L[-1][0])
```

```python
L = [[[1,2], [3,4]], [[5,6], [7,8]]]

print(L[1][0][0])
```

```python
l = [1,2,3,4,5,6]

print(l[0:3])
print(l[-3:])
print(l[::-1])
print(l[0::2])
```

### Adding items to a list

```python
#append()

l = [1,2,3]
l.append(4)
print(l)
```

```python
l = [1,2,3,4]
l.extend([5,6])
print(l)
```

```python
l = [1,2,3]
l.append([4,5])
print(l)
```

```python
l = [1,2,3]
l.extend('hello')
print(l)
```

```python
#insert

l = [1,3,4]
l.insert(1,2)
print(l)
```

### Editing items in a list:

```python
l = [1,2,3,4]
l[0] = 100
print(l)
```

```python
l = [1,2,3,4]
l[0:3] = [100,200,300]
print(l)
```

### Deleting items from a list:

```python
#del()

l = [1,2,3,4]

del l[-1]
print(l)
```

```python
l =[1,2,3,4,5]

del l[2:4]
print(l)
```

```python
#remove()

l = [1,2,3,4]

l.remove(3)
print(l)
```

```python
# pop()

l = [1,2,3,4]

l.pop(0)
print(l)
```

```python
l.pop()
```

```python
#clear()
l = [1,2,3]
l.clear()
print(l)
```

### Operations in list

```python
#arithematic:

l1 = [1,2,3]
l2 = [4,5,6]

print(l1+l2)
```

```python
#membership:

l1 = [1,2,3,[5,6]]

print(3 in l1)
print(4 in l1)

print (5 in l1[3])
```

```python
# iteration

l = [1,2,3,[4,5,6]]

for i in l:
    print(i)
```

```python
L = [[[1,2], [3,4]], [[5,6], [7,8]]]

for i in L:
    print(i)
```

### List comprehension

```python
# Add 1 to 11 numbers in a list

l = [i for i in range(1,11)]
print(l)
```

```python
#Scalar multipliaction on a vector

v = [2,3,4]
s = 3

print(v*s)
```

```python
# languages starting with p only:

lang = ['java', 'c', 'python', 'php', 'javascript']

print([i for i in lang if i[0]=='p'])
```
