#### Aggregation

```python
class Customer:

    def __init__(self,name,gender,address):
        self.name = name
        self.gender = gender
        self.address = address

    def print_add(self):
        print(self.address.get_city(), self.address.state, self.address.pin)

    def edit_profile(self,new_name, new_city, new_state, new_pin):
        self.new_name = new_name
        self.address.edit_address(new_city, new_state, new_pin)

class Address:

    def __init__(self,city,pin,state):
        self.__city = city
        self.pin = pin
        self.state = state

    def get_city(self):
        return self.__city
    

    def edit_address(self,new_city, new_state, new_pin):
        self.__city = new_city
        self.state = new_state
        self.pin = new_pin

add1 = Address('gurgaon', 122011, 'haryana')
cust = Customer('shubh', 'male', add1)

cust.print_add()

cust.edit_profile('ankit', 'mumbai', 'maharashtra', 11111)

cust.print_add()
```

#### Method Overriding

#### Super Keyword

```python
class Phone:
    def __init__(self, price, brand, camera):
        print('Inside phone constructor')
        self.price = price
        self. brand = brand
        self.camera = camera

    def buy(self):
        print('Buying a phone')


class Smartphone(Phone):
    def buy(self):
        print('Buying a smartphone')
        super().buy()


s = Smartphone(2000, 'Apple', 13)
s.buy()
```

```python
# super -> contructor:

class Phone:
    def __init__(self, price, brand, camera):
        print("inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera


class Smartphone(Phone):
    def __init__(self, price, brand, camera, os, ram):
        print('Inside Smartphone constructor')
        super().__init__(price, brand, camera)

        self.os = os
        self. ram = ram
        print("inisde smartphone construtor")

s = Smartphone(2000, 'samsung', 12, 'Android', 2)

print(s.os)
print(s.brand)
```

##### Multilevel Inheritance

```python
class Product:
    def review(self):
        print('Product customer review')

class Phone(Product):
    def __init__(self, price, brand):
        print('Inside Phone Constructor')
        self.__price = price
        self.brand = brand

    def buy(self):
        print('Buying a phone')


class Smartphone(Phone):
    pass

s=Smartphone(200, 'Apple')
s.buy()
s.review()
```

#### Hierarchical Inheritance

```python
class Phone:
    def buy(self):
        print('Buying a phone')

class Smartphone(Phone):
    def buy(self):
        print('Buying a smartphone')

class Featurephone(Phone):
    def buy(self):
        print('Buying a feature phone')


s = Smartphone()
f = Featurephone()
```

## Polymorphism : Method Overriding, Method Overloading, Operator Overloading

```python
#Method Overloading

class Shape:
    def area(self, r):
        return 3.14*r*r
    
    def area(self, l,b):
        return l*b
    
s = Shape()

s.area(5,7)
```

##### Python doesnt have method overloading

```python
#Method overloading using default parameters

class Shape():
    
    def area(self, a, b=0):

        if b==0:
            return 3.14*a*a
        
        else:
            return a*b
        
s = Shape()
print(s.area(5))
print(s.area(5,6))
```
