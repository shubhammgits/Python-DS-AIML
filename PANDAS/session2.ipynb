# Data Frame

### Creating DataFrame

```python
import numpy as np
import pandas as pd
```

```python
#Using Lists (2D):

student_data = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]

pd.DataFrame(student_data, columns = ['iq', 'marks', 'package'])
```

```python
# using dict:

student_dict = {
    'iq' : [100,90,120,80],
    'marks' : [90,70,100,50],
    'package' : [10,7,14,2]
}

stud_df = pd.DataFrame(student_dict)
print(stud_df)
```
