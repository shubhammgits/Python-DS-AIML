# Data Frame Functions

```python
import numpy as np
import pandas as pd
```

### value_counts() [series and dataframe]

```python
#series
a = pd.Series([1,2,3,1,2,3,1])
print(a)
```

```python
a.value_counts()
```

```python
#dataframe:

marks = pd.DataFrame([
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,70,14],
    [80,70,14]
], columns = ['iq', 'marks', 'package'])

print(marks)
```

```python
marks.value_counts()
```
