# Data Frame Functions

```python
import numpy as np
import pandas as pd
```

### value_counts() [series and dataframe]

```python
#series
a = pd.Series([1,2,3,1,2,3,1])
print(a)
```

```python
a.value_counts()
```

```python
#dataframe:

marks = pd.DataFrame([
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,70,14],
    [80,70,14]
], columns = ['iq', 'marks', 'package'])

print(marks)
```

```python
marks.value_counts()
```

```python
ipl = pd.read_csv('datasets\ipl matches.csv')
ipl.head(2)
```

```python
# find which player has won the most potm --> in finals and qualifiers

ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```

```python
#toss decision plot:
ipl['TossDecision'].value_counts().plot(kind = 'pie')
```

```python
#how many matches each team has played

(ipl['Team2'].value_counts() + ipl['Team1'].value_counts()).sort_values(ascending=False)
```

```python
#sorting multiple columns:

stud = pd.DataFrame({
    'name' : ['nitish', 'ankit', 'rupesh', np.nan, 'mrityunjay', np.nan,'rishabbh',np.nan, 'aditya', np.nan],
    'college' : ['bit', 'iit', 'vit', np.nan, np.nan, 'vlsi', 'ssit', np.nan, np.nan, 'git'],
    'branch' : ['eee', 'it', 'cse', np.nan, 'me', 'ce', 'civ', 'cse','bio', np.nan],
    'cgpa' : [6.66,8.25,6.41,np.nan,5.6,9.0,7.4,10,7.4,np.nan],
    'package' : [4,5,6,np.nan,6,7,8,9,np.nan,np.nan]
})

stud
```

```python
stud.sort_values('name')
```

```python
stud.sort_values('name', na_position='first')
```
