# GroupBy Object In Pandas

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('datasets/imdb-top-1000.csv')
```

```python
movies.head(2)
```

```python
genre = movies.groupby('Genre')
print(genre)
```

```python
#apllying builtin aggregate functions on groupby object
```

```python
genre.sum()
```

```python
genre.sum('Director')
```

```python
genre.max()
```

```python
genre.min('Director')
```

```python
genre.mean('Director')
```

```python
#Find the top 3 genres by total earnings:

movies.groupby('Genre').sum()['Gross'].sort_values(ascending = False).head(3)
```

```python
movies.groupby('Genre')['Gross'].sum()
```

```python
#Find the genre with the highest avg in IMDB Rating:

movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
#find the director with the most popularity:

movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending = False).head(1)
```
