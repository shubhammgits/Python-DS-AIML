# Tuples

### Creation of tuple

```python
t1 = ()
type(t1)
```

```python
t1 = (2,)
t2 = ('hello',)
print(type(t1))
print(type(t2))
```

```python
t3 = (1,2,3,4) #hetro
t4 = (1,True, [1,2,3], 'hello')

print(t3)
print(t4)
```

```python
t5 = (1,2,3,('hello', True)) #2d

print(t5)
```

```python
t6 = tuple('hello')
print(t6)
```

```python
a = (1,2,3,4)
a = a + (5,)

print(a)
```

### Accessing Items : Indexing and Slicing

```python
t3 = (1,2,3,4)

print(t3[0])
print(t3[-1])

print(t3[1:3])
print(t3[-1:-4:-1])
print(t3[::-1])
```

##### Tuples are immutable so no editing items, no adding items, no deleting items can be done

### Operations on tuples


```python
#concatenation and multiplication

t1 = (1,2,3,4)
t2 = (5,6,7,8)

print(t1+t2)

print(t1*4)
```

```python
#Membership

t1 = ('hello', 'shubham', 1, True)

print(1 in t1)
print(False in t1)
```

```python
#Iteration

t1 = ('hello', 'shubham', 1, True)

for i in t1:
    print(i, end=' ')
```

### Functions in tuples


```python
#len/sum/min/max/sorted

t1 = (1,2,3,4)

print(len(t1))
print(sum(t1))
print(min(t1))
print(max(t1))
print(sorted(t1))
print(sorted(t1, reverse=True))
```

### Tuples are faster than List

```python
import time

L = list(range(10000000))
T = tuple(range(10000000))

start = time.time()

for i in L:
    i*10
print("List Time = ", time.time() - start)

start = time.time()

for i in T:
    i*10
print('Tuple Time = ', time.time() - start)
```

### Tuples take less memory space than List

```python
import sys

L = list(range(10000000))
T = tuple(range(10000000))

print('List size = ', sys.getsizeof(L))
print('Tuple size = ', sys.getsizeof(T))
```

### Tuple Unpacking

```python
a,b,c = (1,2,3)

print(a,b,c)
```

```python
#Swapping Numbers:

a=1
b=2

a,b = b,a

print(a,b)
```

```python
a,b,*others = (1,2,3,4,5,6)

print(a,b)
print(others)
```

```python
#zipping tupples:

a = (1,2,3,4)
b = (5,6,7,8)

print(zip(a,b))

print(tuple(zip(a,b)))
print(list(zip(a,b)))
```
