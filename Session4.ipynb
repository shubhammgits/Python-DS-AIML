# Tuples

### Creation of tuple

```python
t1 = ()
type(t1)
```

```python
t1 = (2,)
t2 = ('hello',)
print(type(t1))
print(type(t2))
```

```python
t3 = (1,2,3,4) #hetro
t4 = (1,True, [1,2,3], 'hello')

print(t3)
print(t4)
```

```python
t5 = (1,2,3,('hello', True)) #2d

print(t5)
```

```python
t6 = tuple('hello')
print(t6)
```

### Accessing Items : Indexing and Slicing

```python
t3 = (1,2,3,4)

print(t3[0])
print(t3[-1])

print(t3[1:3])
print(t3[-1:-4:-1])
print(t3[::-1])
```

##### Tuples are immutable so no editing items, no adding items, no deleting items can be done

### Operations on tuples


```python
#concatenation and multiplication

t1 = (1,2,3,4)
t2 = (5,6,7,8)

print(t1+t2)

print(t1*4)
```

```python
#Membership

t1 = ('hello', 'shubham', 1, True)

print(1 in t1)
print(False in t1)
```

```python
#Iteration

t1 = ('hello', 'shubham', 1, True)

for i in t1:
    print(i, end=' ')
```

### Functions in tuples


```python
#len/sum/min/max/sorted

t1 = (1,2,3,4)

print(len(t1))
print(sum(t1))
print(min(t1))
print(max(t1))
print(sorted(t1))
print(sorted(t1, reverse=True))
```
