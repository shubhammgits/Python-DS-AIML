# Tuples

### Creation of tuple

```python
t1 = ()
type(t1)
```

```python
t1 = (2,)
t2 = ('hello',)
print(type(t1))
print(type(t2))
```

```python
t3 = (1,2,3,4) #hetro
t4 = (1,True, [1,2,3], 'hello')

print(t3)
print(t4)
```

```python
t5 = (1,2,3,('hello', True)) #2d

print(t5)
```

```python
t6 = tuple('hello')
print(t6)
```

```python
a = (1,2,3,4)
a = a + (5,)

print(a)
```

### Accessing Items : Indexing and Slicing

```python
t3 = (1,2,3,4)

print(t3[0])
print(t3[-1])

print(t3[1:3])
print(t3[-1:-4:-1])
print(t3[::-1])
```

##### Tuples are immutable so no editing items, no adding items, no deleting items can be done

### Operations on tuples


```python
#concatenation and multiplication

t1 = (1,2,3,4)
t2 = (5,6,7,8)

print(t1+t2)

print(t1*4)
```

```python
#Membership

t1 = ('hello', 'shubham', 1, True)

print(1 in t1)
print(False in t1)
```

```python
#Iteration

t1 = ('hello', 'shubham', 1, True)

for i in t1:
    print(i, end=' ')
```

### Functions in tuples


```python
#len/sum/min/max/sorted

t1 = (1,2,3,4)

print(len(t1))
print(sum(t1))
print(min(t1))
print(max(t1))
print(sorted(t1))
print(sorted(t1, reverse=True))
```

### Tuples are faster than List

```python
import time

L = list(range(10000000))
T = tuple(range(10000000))

start = time.time()

for i in L:
    i*10
print("List Time = ", time.time() - start)

start = time.time()

for i in T:
    i*10
print('Tuple Time = ', time.time() - start)
```

### Tuples take less memory space than List

```python
import sys

L = list(range(10000000))
T = tuple(range(10000000))

print('List size = ', sys.getsizeof(L))
print('Tuple size = ', sys.getsizeof(T))
```

### Tuple Unpacking

```python
a,b,c = (1,2,3)

print(a,b,c)
```

```python
#Swapping Numbers:

a=1
b=2

a,b = b,a

print(a,b)
```

```python
a,b,*others = (1,2,3,4,5,6)

print(a,b)
print(others)
```

```python
#zipping tupples:

a = (1,2,3,4)
b = (5,6,7,8)

print(zip(a,b))

print(tuple(zip(a,b)))
print(list(zip(a,b)))
```

### Operations on sets

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

print('Union = ', s1 | s2)
print("intersection : ", s1 & s2)
print('Difference : ', s1-s2)
print('Symmetric Difference : ', s1^s2)

print(1 in s1)
print(10 in s2)


for i in s1:
    print(i, end=' ')
```

### Functions on sets

```python
s1 = {1,2,3,4}

print('Length  = ', len(s1))
print('Max = ', max(s1))
print('Min = ', min(s1))
print('Sum = ',sum(s1))

print("Sorted = ", sorted(s1))
print('Reverse Sorted = ', sorted(s1, reverse=True))
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.union(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

print(s1.difference(s2))

s1.difference_update(s2)

print(s1)
print(s2)

```

```python
s1 = {1,2,3,4}
s2 = {3,4,5,6}

s1.isdisjoint(s2)
```

### 2D dictionary

```python
stud = {
    'name' : 'Shubham',
    'gender' : 'male',
    'age' : 21,
    'subjects' : {
        'dsa' : 50,
        'eng' : 60,
        'maths' : 74
    }
}
```

### Accessing Items

```python
for i in stud:
    print(i,stud[i])

stud['subjects']['dsa']
```

### Adding new key value pair:

```python
stud['Sem'] = 6
stud['subjects']['toc'] = 65
print(stud)
```

### Remove key-value pair

```python
#pop()

stud.pop('age')
print(stud)
```

```python
#popitems()

stud.popitem()
print(stud)
```

```python
#del

del stud['subjects']['eng']
print(stud)
```

```python
#clear

stud.clear()
print(stud)
```

### Editing Key-value pairs

```python
stud = {
    'name' : 'Shubham',
    'gender' : 'male',
    'age' : 21,
    'subjects' : {
        'dsa' : 50,
        'eng' : 60,
        'maths' : 74
    }
}

stud['name'] = 'shubham'
print(stud)
```

### Operations

```python
# Membership

'name' in stud
```

```python
# iteration

for i in stud:
    print(i)
```

### Functions

```python
#len/min/max/sorted

print(stud)

print(len(stud))

print(min(stud))
print(max(stud))
print(sorted(stud))
print(sorted(stud, reverse=True))
```

```python
#items/keys/values

print(stud.items())


print(stud.keys())


print(stud.values())
```

```python
# update

d1 = {1:2 , 3:4, 5:7}
d2 = {5:6}

d1.update(d2)

print(d1)
```

### Dictionary Comprehension

```python
{i:i**2 for i in range(1,11)}
```

```python
#using existing dictionary :

distances = {'delhi' : 1000 , 'mumbai' : 2000, 'banglore' : 3000}

print(distances.items())

{key:value*0.62 for (key,value) in distances.items()}
```

```python
#using zip

days = ('Sunday' , 'Monday' , 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday')

temp = (30.5, 32.6, 31.8, 33.4, 29.8, 30.2, 29.9)

{i:j for (i,j) in zip(days,temp)}
```

```python
# using if condition

products = {'phone' : 10, 'laptop' : 0, 'charger' : 5, 'tablet' : 0}

{i:j for (i,j) in products.items() if j>0}
```
