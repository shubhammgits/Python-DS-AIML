## Types of Arguments

## *args and *kwargs

```python
#*args:

def multiply(*args):

    prod = 1

    for i in args:
        prod *= i

    return prod

multiply(1,2,3,4,4,5,5)
```

```python
#*kwargs

def display(**kwargs):

    for (key,value) in kwargs.items():
        print(key,':',value)


display(India = 'Delhi', Pakistan = 'Islamabad', Srilanka = 'Colombo', Nepal = 'Kathmandu')
```

## Nested Function

```python
def f():
    def g():
        print('Inside function g')
    g()
    print('Inside function f')

f()
```

```python
def g(x):
    def h():
        x='abc'
    x=x+1
    print('in g(x) : x = ',x)

    h()
    return x

x=3
z=g(x)

print(z)
```

```python
def g(x):
    def h(x):
        x=x+1
        print('in h(x) : x = ',x)

    x=x+1
    print('in g(x) : x = ',x)
    h(x)
    return x

x=3
z=g(x)
print('In main program scope : x = ',x)
print('In main program scope : z = ',z)
```

```python
def g(x):
    def h(x):
        x = x+1
        print('In h(x) : x = ',x)

    x=x+1
    print('In g(x) : x = ',x)
    h(x)
    return x

x=3
z=g(x)
print('In main program scope : x = ',x)
print('In main program scope : z = ',z)
```

## Lambda Function

## Higher Order Function

```python
def square(x):
    return x**2


def transform(f,L):
    output = []

    for i in L:
        output.append(f(i))

    print(output)


L = [1,2,3,4,5]
transform(square, L)
```

```python
# using lambda:

def transform(f1, f2, L):
    output1=[]
    output2=[]

    for i in L:
        output1.append(f1(i))
        output2.append(f2(i))

    print(output1)
    print(output2)


L = [1,2,3,4,5]
transform(lambda x:x**2, lambda x:x**3, L)
```
