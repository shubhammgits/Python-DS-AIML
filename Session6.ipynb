## Creating a function with docstring

```python
def is_even(num):
    """
    This function returns if given number is odd or even
    input - any valid ineteger
    output - odd/even
    created on - 14/01/2026
    created by - Shubham
    """


    if num%2==0:
        return 'even'
    else:
        return 'odd'
```

```python
#Calling the function:

for i in range(1,11):
    x = is_even(i)
    print(i,':',x)
```

```python
# Correct way:

def is_even(num):
    """
    DOCSTRING
    """


    if type(num) == int:
        if num%2==0:
            return 'even'
        else:
            return 'odd'
        
    else:
        return 'Pagal hai kya?'
```

```python
# calling the functions :

is_even('hello')
```

## Types of Arguments

```python
# Default argument

def power(a=1, b=1):
    return a**b

print(power(2,3))
print(power(2))
print(power())
```

```python
# Positional argument

power(2,3)    #a=2, b=3
```

```python
#keyword argument:

power(b=2, a=3)
```

## *args and *kwargs

```python
#*args:

def multiply(*args):

    prod = 1

    for i in args:
        prod *= i

    return prod

multiply(1,2,3,4,4,5,5)
```

```python
#*kwargs

def display(**kwargs):

    for (key,value) in kwargs.items():
        print(key,':',value)


display(India = 'Delhi', Pakistan = 'Islamabad', Srilanka = 'Colombo', Nepal = 'Kathmandu')
```
