```python
#raise:

class Bank:
    def __init__(self, balance):
        self.balance = balance
    
    def withdraw(self, amount):

        if amount < 0:
            raise Exception('Amount cannot be negative!')
        
        if amount > self.balance:
            raise Exception('Paise nahi hai tere paas!')
        
        self.balance -= amount


obj = Bank(10000)

try:
    obj.withdraw(4000)

except Exception as e:
    print(e)

else:
    print('Balance : ', obj.balance)
```

```python
#custom error class:

class MyException(Exception):
    def __init__(self,message):
        print(message)  

class Bank:
    def __init__(self, balance):
        self.balance = balance
    
    def withdraw(self, amount):

        if amount < 0:
            raise MyException('Amount cannot be negative!')
        
        if amount > self.balance:
            raise MyException('Paise nahi hai tere paas!')
        
        self.balance -= amount


obj = Bank(10000)

try:
    obj.withdraw(-4000)

except MyException as e:
    pass

else:
    print('Balance : ', obj.balance)
```

```python
#Google registration class: custom class functionality!

class SecurityError(Exception):
    def __init__(self, message):
        print(message)

    def logout(self):
        print('logged out from all the devices!')

class Google:
    
    def __init__(self,name,email,password,device):
        self.name = name
        self.email = email
        self.password = password
        self.device = device

    def login(self,email,password,device):

        if device != self.device:
            raise SecurityError('You are hacked!')
        
        if email == self.email and password == self.password:
            print('Welcome!')

        else:
            print('Login error!')
obj = Google('Shubham', 'shubhk918@gmail.com', '1234', 'android')

try:
    obj.login('shubhk918@gmail.com', '1234', 'windows')

except SecurityError as e:
    e.logout()

else:
    print(obj.name)

finally:
    print('Database connection closed')
```
