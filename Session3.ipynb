```python
#Sequence Sum: 1/1! + 2/2! + 3/3! +.....+N/N!


n = int(input("Enter the Nth Number : "))
fact, sum = 1,0

for i in range(1, n+1):
    fact = fact*i
    seq = i/fact
    sum = sum + seq
 
print(sum)
```

# Patterns

#### Pattern 1

```python
rows = int(input('Enter the number of rows : '))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print('*', end='')
    print()
```

#### Pattern 2

```python
for i in range(1,6):
    for j in range(1,i+1):
        print(j, end = '')
    print()
```

#### Pattern 3

```python
for i in range(1, 5):
    for j in range(1,i+1):
        print(j, end='')
    for k in range (i-1, 0, -1):
        print(k, end='')
    
    print()
```

```python
#Find the length of a given string without using len() functions

s = input('Enter a string : ')

counter = 0

for i in s:
    counter+=1

print('Length of string is ',counter)
```

```python
#Extract username from email

email = input('Enter your email : ')

for i in email:
    if(i=='@'):
        break
    print(i, end='')
```

```python
#or

email = input('Enter email : ')
index = email.index('@')
print(email[0:index])
```

```python
#Count the frequency of the character in a string

str = 'hello how are you, Shubham?'
ch = 'h'
counter = 0
for i in str:
    if(i == ch):
        counter+=1

print('Frequency of',ch,'is = ',counter)
```

```python
# Remove a particular character from a string

str = 'hello, Shubham!'
ch = 'h'

result=''

for i in str:
    if i!=ch:
        result = result + i

print(result)
```

```python
#Write a program that can check whether a given strig is palindrome or not!

s = input('Enter a string : ')
flag = True

for i in range(0, len(s)//2):
    if s[i] != s[len(s) - i - 1]:
        flag = False
        print('Not a Palindrome')
        break

if flag:
    print('Palindrome!')
```

```python
# Write a program to count the number of words in a string without split()

s = 'hi how are you'
temp = ''
L = []

for i in s:
    if i!=' ':
        temp = temp+i
    else:
        L.append(temp)
        temp = ''

L.append(temp)
print(L)
```

```python
# Write a python program to convert a string to title case without using title()

s = 'hI hOW Are yOu'
L=[]

for i in s.split():
    L.append(i[0].upper() + i[1:].lower())

print(' '.join(L))
```

```python
#Write a program to convert an integer to string without using str()

number = int(input('Enter the number : '))

digits = '0123456789'
result =''
while number!=0 :

    result = digits[number%10] + result
    number = number//10

print(result)
print(type(result))

```
