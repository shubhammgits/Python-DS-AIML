# Question

```python
#Sequence Sum: 1/1! + 2/2! + 3/3! +.....+N/N!


n = int(input("Enter the Nth Number : "))
fact, sum = 1,0

for i in range(1, n+1):
    fact = fact*i
    seq = i/fact
    sum = sum + seq
 
print(sum)
```

# Patterns

#### Pattern 1

```python
rows = int(input('Enter the number of rows : '))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print('*', end='')
    print()
```

#### Pattern 2

```python
for i in range(1,6):
    for j in range(1,i+1):
        print(j, end = '')
    print()
```

#### Pattern 3

```python
for i in range(1, 5):
    for j in range(1,i+1):
        print(j, end='')
    for k in range (i-1, 0, -1):
        print(k, end='')
    
    print()
```

# Loop Control Statement

#### Break

```python
for i in range(1,10):
    if i==5:
        break
    print(i)
```

### Question

```python
# Prime numbers between a range:
import math
for i in range(10,100):
    for j in range(2,int(math.sqrt(i))+1):
        if i%j == 0:
            break
    else:
        print(i)
```

#### Continue

```python
for i in range(1,10):
    if i==5:
        continue
    print(i)
```

# Strings

```python
# Creation of string:

s = 'hello'
s = "hello"
s = '''hello'''
s = """hello"""

s = str(8)

line = "It's raining outside!"

print(s)
print(line)
```

```python
#Accessing Substrings from a string

#INDEXING

s = 'hello'

print(s[0])
print(s[1])
print(s[2])
```

```python
#Negative Indexing

s = 'hello world'
print(s[-1])
print(s[-2])
print(s[-3])
print(s[-4])
```

```python
#Slicing

s = 'hello world'

print(s[0:5])
print(s[1:5])
print(s[3:5])


print(s[2:])
print(s[:3])
```

```python
#Step size:

s = 'hello world'

print(s[0:8:2])
print(s[2:8:1])
print(s[6:0:-1])
```

```python
# Reverse a string:

s = 'hello world'
print(s[::-1])
```

```python
s = 'hello world'
print(s[-5:])
print(s[-1:-6:-1])
print(s[-1:-5:-2])
```
