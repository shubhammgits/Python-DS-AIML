```python
celc = float(input('Enter the temprature in celcius : '))
fahren = (celc * 9/5)+32
print('Temprature in Fahrenheit : ',fahren)
```

```python
a = int(input('Enter the first number : '))
b = int(input('Enter the second number : '))

print('Before swapping : a = ',a,'b = ',b)
temp = a
a = b
b = temp

print('After Swapping : a = ',a,'b = ',b)
```

```python
p1x = int(input('Enter x coordinate of point 1 : '))
p1y = int(input('Enter y coordinate of point 1 : '))
p2x = int(input('Enter x coordinate of point 2 : '))
p2y = int(input('Enter y coordinate of point 2 : '))

euclid = ((p2x - p1x)**2 + (p2y - p1y)**2)**0.5

print('Euclidean distance between these two points : ',round(euclid,2))
```

```python
p = int(input('Enter the principle amount : '))
r = float(input('Enter the rate : '))
t = int(input('Enter the time period : '))

si = (p*r*t)/100

print('Simple Interest = ',si)
```

```python
n = int(input('Enter the value of n :'))

sum = 0

for i in range(1,n+1):
    sum = sum + i**2

print('Sum of squares of first {} natural numbers is :'.format(n), sum)
```

```python
n1 = int(input('Numerator 1 : '))
d1 = int(input('Denominator 1 : '))
n2 = int(input('Numerator 2 : '))
d2 = int(input('Denominator 2 : '))

ns = (n1*d2) + (n2*d1)
ds = d1*d2

print('Sum of {}/{} and {}/{} is {}/{}'.format(n1,d1,n2,d2,ns,ds))
```
