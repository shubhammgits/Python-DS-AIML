#### Sum of 3 digit number entered by the user

```python
num = int(input('Enter the number : '))

quot1 = num%10
num = num//10

quot2 = num%10
num = num//10

quot3 = num%10

print('Sum = ',quot1+quot2+quot3)
```

#### Sum of n-digit number entered by the user

```python
num = 1234

sum=0

while num>0:
    sum = sum + num%10
    num = num//10

print('Sum = ', sum)
```

#### Guessing Game :

```python
import random

randNum = random.randint(1,100)
userNum = int(input("Guess the number between 1 to 100 : "))
tries = 1

while(userNum != randNum):
    if userNum > randNum : 
        userNum = int(input("Guess smaller number : "))
        tries+=1
    
    else:
        userNum = int(input("Guess larger number : "))
        tries+=1
    
else:
    print('Correct Guess!')
    print('Attempts : ',tries)
```

#### Sum/Factorial sequence : 1/1! + 2/2! + 3/3! +....+N/N!

```python
num = int(input('ENter the nth term : '))

sum,fact =0,1

for i in range(1,num+1):
    fact *= i
    seq = i/fact

    sum = sum+seq

print('Sum = ',sum)
```

#### Patterns


```python
rows = 5

for i in range(1,rows+1):
    for j in range(1,i+1):
        print('*', end='')
    print()
```

```python
for i in range(5,0,-1):
    for j in range(i,0,-1):
        print('*',end='')
    print()
```

```python
for i in range(1,6):
    for j in range(1,i+1):
        print(j,end='')
    for k in range(i-1,0,-1):
        print(k, end='')
    print()
```

```python
for i in range(5,0,-1):
    for j in range(i,0,-1):
        print(j,end='')
    print()
```

```python
for i in range(1,7):
    for j in range(1,i+1):
        print(' ',end='')
    for k in range(1,i+1):
        print('*',end='')
    print()
```
