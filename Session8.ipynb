```python
#OOP class for : a user can create and view 2d coordinates with following functionalities:

# distance between 2 coordiates, distance of a coordinate from origin, chekc if point lies on a given line, user can find the distance between a given 2d point and a given line:
```

```python
class Point:
    def __init__(self,x,y):
        self.x_cord = x
        self.y_cord = y

    def __str__(self):
        return '<{},{}>'.format(self.x_cord,self.y_cord)
    
    def euclidean_distance(self,other):

        return ((self.x_cord - other.x_cord)**2 + (self.y_cord - other.y_cord)**2)**0.5
    
    def dist_origin(self):

        return ((self.x_cord)**2 + (self.y_cord)**2)**0.5
    

class Line:

    def __init__(self,A,B,C):
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return '{}x + {}y + {} = 0'.format(self.A,self.B,self.C)
    
    def point_on_line(line,point):

        if line.A*point.x_cord + line.B*point.y_cord + line.C == 0:
            return 'Point lies on the line'
        
        else:
            return 'Point does not lie on the line'
        

    def shortest_distance(line,point):
        
       return abs(line.A*point.x_cord + line.B*point.y_cord + line.C)/((line.A**2 + line.B**2))**0.5 
    

    def intersection_of_lines(line1,line2):

        if line1.A*line2.B != line1.B*line2.A:
            return 'Lines Intersect!'
        
        else:

            if line1.B*line2.C != line2.B*line1.C:
                return 'Parallel Lines! No Intersection'
            
            else:
                return 'Coincident: Infinite Intersection!'
```

```python
p1 = Point(0,0)
p2 = Point(10,10)
print(p1)
print(p2)


print(p1.euclidean_distance(p2))

print(p2.dist_origin())
```

```python
l1 = Line(1,1,-2)
p1 = Point(1,10)

print(l1.point_on_line(p1))

l1.shortest_distance(p1)
```

```python
l1 = Line(2,3,-5)
l2 = Line(4,-1,1)

l1.intersection_of_lines(l2)
```

```python
l3 = Line(2,3,-6)
l4 = Line(4,6,5)

l3.intersection_of_lines(l4)
```

```python
l5 = Line(2,3,-6)
l6 = Line(4,6,-12)

l5.intersection_of_lines(l6)
```

## Objects have the power to access the attributes of its class

```python
class Person:
    def __init__(self,name_inp,country_inp):
        self.name = name_inp
        self.country = country_inp

    def greet(self):
        if self.country == 'india':
            print("Namaste",self.name)

        else:
            print("Hello",self.name)
```

```python
p = Person('shubham','india')

print(p.name)
print(p.country)

print(p.greet())
```

```python
#If we try to access non-existen attribute:

p.gender
```

```python
#we can add attribute to class from outside the class:

p.gender = 'male'

print(p.gender)
```
