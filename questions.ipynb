#### Sum of 3 digit number entered by the user

```python
num = int(input('Enter the number : '))

quot1 = num%10
num = num//10

quot2 = num%10
num = num//10

quot3 = num%10

print('Sum = ',quot1+quot2+quot3)
```

#### Sum of n-digit number entered by the user

```python
num = 1234

sum=0

while num>0:
    sum = sum + num%10
    num = num//10

print('Sum = ', sum)
```

#### Login program

```python
email = 'abc123@gmail.com'
passw = 1234

userEmail = input('Enter the email : ')
userPassw = int(input('Enter the password : '))

if userEmail == email and userPassw == passw:
    print('Login Successful!')

elif userPassw != passw:
    print('Password Incorrect try again!')
    userPassw = int(input('Enter the password : '))

    if userPassw != passw:
        print('Beta tumse nahi ho payega!')
    
    else:
        print('Welcome!')

else:
    print('Incorrect Credentials!')
```

#### Guessing Game :

```python
import random

randNum = random.randint(1,100)
userNum = int(input("Guess the number between 1 to 100 : "))
tries = 1

while(userNum != randNum):
    if userNum > randNum : 
        userNum = int(input("Guess smaller number : "))
        tries+=1
    
    else:
        userNum = int(input("Guess larger number : "))
        tries+=1
    
else:
    print('Correct Guess!')
    print('Attempts : ',tries)
```
