#### Sum of 3 digit number entered by the user

```python
num = int(input('Enter the number : '))

quot1 = num%10
num = num//10

quot2 = num%10
num = num//10

quot3 = num%10

print('Sum = ',quot1+quot2+quot3)
```

#### Sum of n-digit number entered by the user

```python
num = 1234

sum=0

for i in range(0, len(num)+1):
    sum = sum + num%10
    num = num//10

print('Sum = ', sum)
```
