# Operators

```python
#Arithematic Operators

print(5+5)

print(5-5)

print(2*4)

print(5%3)

print(5/2)

print(5//2)

print(5**2)
```

```python
# Relational Operators

print(4>5)
print(4<5)
print(4!=5)
```

```python
#Logical Operators

print(1 and 0)
print(1 or 0)
print(not 1)


print(True and False)
```

```python
#Bitwise Operator


#Bitwise and (&)
print(2&3)

#Bitwise OR (|)
print(2|3)

#Bitwise XOR (^)
print(8^6)

#Bitwise Not (~)
print(~8)


print(4>>2)
print(5<<2)
```

```python
#Assignment Operator

a=4
a+=2
print(a)

b=2
b-=2
print(b)

a+=1
print(a)


str1 = 'shubham'
str1+='jha'

print(str1)
```

```python
#Membership Operator

print('D' in 'Delhi')
print(5 in [1, 2, 3, 4])
```

# If-else in Python

```python
# Login program and indentation (Let email : abc@gmail.com and pass : 123)


email = input("Enter your email : ")
password = input("Enter your password : ")

corremail = 'abc@gmail.com'
corrpass = '123'

if(email == corremail and password == corrpass) :
    print("Login Succesful!")

else:
    print("Wrong email or password!")
```
