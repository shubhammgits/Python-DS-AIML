# If-else in Python

```python
# Login program and indentation (Let email : abc@gmail.com and pass : 123)


email = input("Enter your email : ")
password = input("Enter your password : ")

corremail = 'abc@gmail.com'
corrpass = '123'

if email == corremail and password == corrpass :
    print("Login Succesful!")

elif email == corremail and password != corrpass :
    print('Incorrect password.')
    password = input("Enter password again : ")

    if password == corrpass :
        print('Welcome!')
    else :
        print('Beta tumse nahi ho payega!')

else:
    print("Wrong email or password!")
```

```python
# Menu driven calculator

a = int(input("Enter the first number : "))
b = int(input("Enter the second number : "))

op = input("Enter the operator : ")

if op == '+' : 
    print(a+b)

elif op == '-' :
    print(a-b)

elif op == '*' :
    print(a*b)

elif op == '/' :
    print(a/b)

elif op == '**' :
    print(a**b)

elif op == '//' :
    print(a//b)

else :
    print("Incorrect Operator!")
```

# LOOPS in Python

```python
# While loop

# Program to print the table of a number 

num = int(input("Enter the number : "))

i = 1
while i<11:
    print(num, '*', i, '=', num*i)
    i+=1 
```

```python
# While loop with else

x=1
while(x<3):
    print(x)
    x+=1

else:
    print('x printed two times')
```

# Guessing Game

```python
import random

randNum = random.randint(1,100)

userNum = int(input("Guess the number between 1 and 100 : "))

tries = 1

while(userNum != randNum):
    if(userNum > randNum):
        print('Guess Smaller Number')

    else:
        print("Guess Larger Number!")

    userNum = int(input('Guess Again!'))
    tries+=1

else:
    print('Correct Guess!')
    print('Attempts : ', tries)
```
